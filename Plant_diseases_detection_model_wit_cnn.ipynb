# Importing neede files

```python
import os
from os import path
import cv2
import Augmentor
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

# Defining The directory's that we will work with

```python
kaggle_train_dir = 'Data/kaggle/train'
kaggle_val_dir = 'Data/kaggle/valid'
kaggle_test_dir = 'Data/kaggle/test'
dell_row_data = 'Data/dell/Main'
dell_train_dir = 'Data/dell/train'
dell_test_dir =  'Data/dell/test'
dell_val_dir =   'Data/dell/valid'
processed_train_dir ='Data/MargedAndProccessedData/train'
processed_test_dir = 'Data/MargedAndProccessedData/test'
processed_val_dir = 'Data/MargedAndProccessedData/valid'

if not path.exists(dell_train_dir):
    os.mkdir(dell_train_dir)
if not path.exists(dell_test_dir):
    os.mkdir(dell_test_dir)
if not path.exists(dell_val_dir):
    os.mkdir(dell_val_dir)
if not path.exists(processed_train_dir):
    os.mkdir(processed_train_dir)
if not path.exists(processed_test_dir):
    os.mkdir(processed_test_dir)
if not path.exists(processed_val_dir):
    os.mkdir(processed_val_dir)
```

# Spliting the Dell Dataset

```python
for dell_class in os.listdir(dell_row_data):
    class_images = os.listdir(path.join(dell_row_data,dell_class))
    # fill the test file by 10 test Examples
    all_class_size = len(class_images)
    val_size = int(all_class_size*0.2)
    test_size = int(all_class_size*0.1)
    train_size = all_class_size-(val_size+test_size)
    print(f'Class {dell_class} have {all_class_size} images ')
    print (f'split Them to :\n{train_size} Train images')
    print (f'{val_size} Validation images')
    print (f'{test_size} Test images')
    for i in range(len(class_images)):
        image = cv2.imread(path.join(dell_row_data,dell_class,class_images[i]))
        re_img = cv2.resize(image,(256,256))
        if i<test_size:
            cv2.imwrite(dell_test_dir+'/'+dell_class+'_'+str(i)+'.JPG',re_img)
        elif i<(val_size+test_size):
            if not path.exists(path.join(dell_val_dir,dell_class)):
                os.mkdir(path.join(dell_val_dir,dell_class))
            cv2.imwrite(dell_val_dir+'/'+dell_class+'/'+dell_class+'_'+str(i)+'.JPG',re_img)
        else:
            if not path.exists(path.join(dell_train_dir,dell_class)):
                os.mkdir(path.join(dell_train_dir,dell_class))
            cv2.imwrite(dell_train_dir+'/'+dell_class+'/'+dell_class+'_'+str(i)+'.JPG',re_img)
```

```python
def show_data_distribution(nums_train):
    plt.figure(figsize=(15,15))
    plt.title('Train data distribution ',fontsize=30)
    plt.xlabel('Number of image',fontsize=20)
    plt.ylabel('Planet',fontsize=20)

    keys = list(nums_train.keys())
    # get values in the same order as keys, and parse percentage values
    vals = list(nums_train.values())
    sns.barplot(y=keys, x=vals)
    plt.savefig('save_as_a_png.png')

show_data_distribution(train_dic)
```
